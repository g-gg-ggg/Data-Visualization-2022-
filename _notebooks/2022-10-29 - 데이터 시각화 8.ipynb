## s로 끝나는 column 선택하기

```python
df.columns
```

```python
df.iloc[:,map(lambda x : 's' == x[-1], df.columns)]
```

## c 혹은 d로 시작하는 column 선택

```python
df.iloc[:,map(lambda x : 'c' == x[0] or 'd' == x[0], df.columns)] # 람다에서 x라고 칭했으면 x에 대해서 표현하기 # or로 표현하기
```

## 열 만들기

```python
df = pd.DataFrame({'a':[1,2,3],'b':[2,3,4]})
df
```

```python
df['c']=[3,4,5]
df['d']=[4,5,6]
df
```

## 열 추가하기 .assign


```python
df = pd.DataFrame({'a':[1,2,3],'b':[2,3,4]})
df
```

```python
df.assign(c = [3, 4, 5])
```

```python
df.assign(c = [3, 4, 5]).assign(d=[4,5,6]) # 앞에 거는 이 전 데이터 프레임으로 돌아가지 못함. 하지만 sssign은 df를 치면 원래 데이터 모습이 나옴
```
