```python
np.random.seed(43052)
x1,y1 =  np.random.multivariate_normal([-2,-2],[[1,-0.8],[-0.8,1]],size=500).T
x2,y2 =  np.random.multivariate_normal([2,2],[[1,-0.7],[-0.7,1]],size=500).T
```

```python
fig,ax = plt.subplots(2,2)
```

```python
ax[0,0].plot(x1,y1,'o', color = "C0")
ax[0,0].plot(x2,y2,'o', color = "C1", alpha = 0.1)
ax[0,0].set_title('(a) matplotlib - highlight(x1,y1)')

sns.scatterplot(x=x1,y=y1,color = "C0", ax=ax[0,1])
sns.scatterplot(x=x2,y=y2,color = "C1", alpha = 0.1, ax=ax[0,1])
ax[0,1].set_title('(b) seaborn - highlight(x1,y1)')

ax[1,0].plot(x1,y1,'o', color = "C0", alpha = 0.1)
ax[1,0].plot(x2,y2,'o', color = "C1")
ax[1,0].set_title('(c) matplotlib - highlight(x2,y2)')

sns.scatterplot(x=x1,y=y1,color = "C0",alpha = 0.1, ax=ax[1,1])
sns.scatterplot(x=x2,y=y2,color = "C1", ax=ax[1,1])
ax[1,1].set_title('(d) seaborn - highlight(x2,y2)')

fig.tight_layout()
fig
```

```python
#(2) plotnine을 이용하여 아래와 같이 시각화하라.
dd = np.random.seed(43052)
x1,y1 =  np.random.multivariate_normal([-2,-2],[[1,-0.8],[-0.8,1]],size=500).T
x2,y2 =  np.random.multivariate_normal([2,2],[[1,-0.7],[-0.7,1]],size=500).T
```

```python
fig = ggplot(dd)
ggplot(data=dd, mapping=aes(x='x1',y='y1')) + geom_point(mapping=aes(color='cat'),alpha=0.3)
a1 = aes(x = "x1", y = "y1") + a2 = aes(x2 = "x2'", y2 = "y2") 
```

```python
point1 = geom_point(a1, alpha=0.1)
```

```python
line1 = geom_smooth(a1)
```

```python
df = df.set_axis(pd.Index(map(lambda x: x.replace(' ','_'), df.columns)), axis=1)
df.head()
```

```python
#step1: 결측치가 가장 많은 2개의 컬럼을 찾고 이를 제거하라.

df.isna().sum().sort_values(ascending=False)
```

```python
df.drop(columns=['Marking', 'Loaned_From']).dropna()
```

```python
df.Potential
```

```python
#step3: Potential2 = Potential - Overall 를 이용하여 Potential2를 구하라.
df['Potential2'] = df.Potential - df.Overall
```

```python
#
df.groupby(["Age"])['Potential2'].describe()
```

```python
#[0,20,22,26,100]
```

```python
def f(x):
    if x<20: 
        y='~20'
    elif x<22:
        y='20~22'
    elif x<26:
        y='22~26'
    else:
        y='26~' 
    return y
```

```python
df.eval('Potential2 = Potential - Overall').query('Potential2 > 1')\
.assign(Overall_grouped= lambda df: list(map(f,df.Age)))\
.assign(Overall_x= lambda df: list(map(g,df.Overall_grouped)))
```
