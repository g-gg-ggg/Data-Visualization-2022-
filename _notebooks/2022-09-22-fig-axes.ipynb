## 문제1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axs = plt.subplots(2, 3)
```

```python
x,y = [1,2,3,4], [1,2,4,3] 

axs[0,0].plot(x, y,'or')
axs[1,0].plot(x, y,'--or')
axs[0,1].plot(x, y,'og')
axs[1,1].plot(x, y,'--og')
axs[0,2].plot(x, y,'ob')
axs[1,2].plot(x, y,'--ob')

fig
```

## 문제2

```python
fig = plt.figure()
ax1 = fig.add_axes([0.5,0.5,1,1]) # (0,0)의 위치에 (1,1)인 액시즈(=네모틀)을 만들어라. 
ax2 = fig.add_axes([1.0,1.0,1,1])
ax3 = fig.add_axes([1.5,1.5,1,1])

ax1,ax2,ax3 = fig.axes

fig.axes
```

```python
x,y = [1,2,3,4], [1,2,1,1]
ax1.plot(x, y, "or")
ax2.plot(x, y, "og")
ax3.plot(x, y, "ob")

fig
```

## 문제 3

```python
x = np.arange(-5,5,0.1)
y1 = np.sin(x)
y2 = np.sin(2*x) + 2
y3 = np.sin(4*x) + 4 
y4 = np.sin(8*x) + 6
```

```python
plt.plot(x, y1, "--r")
plt.plot(x, y2, "--g")
plt.plot(x, y3, "--b")
plt.plot(x, y4, "--m")
```
