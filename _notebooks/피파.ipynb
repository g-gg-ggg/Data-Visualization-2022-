# float형으로 수정하기 ex)€206.9M ---- value, wage ...

```python
df=pd.read_csv('https://raw.githubusercontent.com/guebin/2021DV/master/_notebooks/2021-10-25-FIFA22_official_data.csv')
```

```python
def convert_currency(value):
    floatvalue = 0.0
    strvalue=""
    if "M" in value:
        strvalue=value.replace("M","").replace("€","")
        floatvalue=float(float(strvalue)*1000000)
    elif "K" in value:
        strvalue=value.replace("K","").replace("€","")
        floatvalue=float(float(strvalue)*1000)
    else:
        floatvalue=value.replace("€","")
    return floatvalue
```

```python
#그냥 위 함수를 적용시키려면 결측치로 인한 오류가 가득 뜨므로 일단 결측치 드랍해서 저장하기
df.dropna(subset=["Release Clause"], how='all', inplace=True)
```

```python
#이제 결측치가 제거된 열에서 함수 적용하기 
df["ReleaseValue"] = df["Release Clause"].apply(convert_currency).astype(float)
```

```python
df["Wage"] = df["Wage"].apply(convert_currency).astype(float)
df["Value"] = df["Value"].apply(convert_currency).astype(float)
```
