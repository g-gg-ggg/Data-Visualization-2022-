```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

**아래의 그림에 대응하는 그림을 seaborn을 이용하여 그려라**

```python
y1 = np.random.randn(90).cumsum() 
y2 = np.random.randn(120).cumsum()
```

```python
plt.plot(y1,'--o')
plt.plot(y2,'--o')
```

```python
df = pd.DataFrame({'index' : list(range(90))+list(range(120)), 'val' : np.concatenate([y1,y2]), 'cat' : ["y1"]*90+["y2"]*120}) 
```

```python
df
```

```python
sns.lineplot(data = df, x = "index", y= "val", hue = "cat")
```

```python
sns.lineplot(data = df, x ="index", y = "val", style = "cat", hue = "cat", dashes = [(5,1),(9,3)], markers = ["o","o"], markersize = 9)
```
